```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(dataset['Thousands of Passengers'], freq=12)
decomposition.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(dataset['Thousands of Passengers'])
```

```python
def adf_check(time_series):
    result = adfuller(time_series)
    print('Augmented Dickey Fuller Test')
    labels = ['ADF Test Statistic', 'P-value', ' #Lags used', ' No of obs']
    
    for value, label in zip(result, labels):
        print(label + ' : '+str(value))
        
    if result[1]<=0.05:
        print('Strong evidence against null hypothesis and my time series is stationary')
    else:
        print('weak evidence against null hypothesis and my time series is non stationary')
```

```python
adf_check(dataset['Thousands of Passengers'])
```

```python
dataset['Thousands of Passengers First Diff'] = dataset['Thousands of Passengers'] - dataset['Thousands of Passengers'].shift(1)
```

```python
dataset
```

```python
adf_check(dataset['Thousands of Passengers First Diff'].dropna())
```

```python
dataset['Thousands of Passengers First Diff'].plot()
```

```python
dataset['Thousands of Passengers 2nd Diff'] = dataset['Thousands of Passengers First Diff'] - dataset['Thousands of Passengers First Diff'].shift(1)
```

```python
dataset
```

```python
adf_check(dataset['Thousands of Passengers 2nd Diff'].dropna())
```

```python
dataset['Thousands of Passengers 2nd Diff'].plot()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
import statsmodels.api as sm
```

```python
model = sm.tsa.statespace.SARIMAX(dataset['Thousands of Passengers'], order=(0,2,0), seasonal_order=(1,2,1,12))
```

```python
results = model.fit()
```

```python
print(results.summary())
```

```python
dataset['forecast'] = results.predict(start=130, end = 144, dynamic=True)
dataset[['Thousands of Passengers', 'forecast']].plot()
```
